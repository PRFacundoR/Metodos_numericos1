```python
#ejericicio 7 metodo normal
import math as m
def funcion(x):
    return 1.01*m.exp(4*x)-4.62*m.exp(3*x)-3.11*m.exp(2*x)+12.2*m.exp(x)-1.99
    
def redondear_sig(x, cifras):
    return float(f"{x:.{cifras - 1}e}")

def truncar_sig(n, cifras):
    if n == 0:
        return 0.0

    signo = -1 if n < 0 else 1
    n = abs(n)
    exp = int(m.floor(m.log10(n)))
    factor = 10**(exp - cifras + 1)
    truncado = m.floor(n / factor) * factor
    return signo * truncado




print("El numero normal es: ", funcion(1.53))
print("El numero truncado es: ",truncar_sig(funcion(1.53),3))
print("El numero redondeado es: ",redondear_sig(funcion(1.53),3))

#metodo horner

def funcionHorner(x):
    e=m.exp(x)
    return e*(e*(e*(e*(1.01)-4.62)-3.11)+12.2)-1.99


print("El numeroH normal es: ", funcionHorner(1.53))
print("El numeroH truncado es: ",truncar_sig(funcionHorner(1.53),3))
print("El numeroH redondeado es: ",redondear_sig(funcionHorner(1.53),3))
```

```python
# ejercicio 5

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.999, 1.001, 1000)

def f(x):
    return x**8 - 8*x**7 + 28*x**6 - 56*x**5 + 70*x**4 - 56*x**3 + 28*x**2 - 8*x + 1
def g(x):
    return (x-1)**8


# Evaluar las funciones
y_f = f(x)
y_g = g(x)

# ----------- Gráfico de f(x) -------------
plt.figure(figsize=(8, 5))
plt.plot(x, y_f, color='blue')
plt.title('Gráfico de f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------- Gráfico de g(x) -------------
plt.figure(figsize=(8, 5))
plt.plot(x, y_g, color='red')
plt.title('Gráfico de g(x)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Los resultados obtenidos no son correctos ya que al analizar las graficas, se ve que ambas son diferentes a pesar de ser la misma funcion algebraicamente y otro error es la cantidad de veces que el polinomio es 0 que es mas de una vez y con numeros erroneos, donde el unico y correcto seria x=1. Estos errores se debe a que f(x) presenta muchas operaciones y en cada operacion se hace redondeo o trucamientos, encambio g(x) a la hora de graficarla se cometen menos errores ya que es solo 2 operaciones que se hacen.
