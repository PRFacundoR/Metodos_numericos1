```python
import numpy as np

def f(x):
    return (1/x) - np.tan(x)  # Podés cambiar esto por cualquier otra función

def secante(a, b, eps, max_iter):
    iter = 0
    xn_1 = a
    xn = b
    xn1 = 2 * xn  # Paso 1

    while abs(xn1 - xn) > eps and iter <= max_iter:  # Paso 2
        f_xn = f(xn)
        f_xn1 = f(xn_1)
        
        if f_xn - f_xn1 == 0:
            print("Error: División por cero.")
            return None
        
        xn1 = xn - f_xn * (xn - xn_1) / (f_xn - f_xn1)
        xn_1 = xn
        xn = xn1
        iter += 1

    if iter > max_iter:  # Paso 3
        print("No converge en", max_iter, "iteraciones.")
    else:
        print("Raíz aproximada =", xn1)

# Ejemplo de uso:
secante(a=0.5, b=1.5, eps=1e-8, max_iter=20)
```

```python


# Definimos el Método de Newton-Raphson
def metodoN_R(f1, f_prime, x0, maX, espa):
    it = 1  # Iniciamos el contador de iteraciones

    while it <= maX and np.abs(f1(x0)) > espa:
        # Fórmula de Newton-Raphson
        x1 = x0 - f1(x0) / f_prime(x0)
        print(f"Iteración {it}: x1 = {x1}, f(x1) = {f1(x1)}")
        x0 = x1  # Actualizamos el valor de x0
        it += 1

    # Resultado final
    if it > maX:
        print(f"No converge en {maX} iteraciones.")
    else:
        print(f"Raíz aproximada: {x1}")

# Definimos una función y su derivada como ejemplos
def f1(x):
    return x**3 - x  # Ejemplo de función

def f_prime(x):
    return 3*x**2 - 1  # Derivada de la función f1(x)





# Parámetros de entrada
x0 = np.sqrt(1/5)  # Valor inicial
maX = 4            # Máximo número de iteraciones
espa = 0.0001      # Tolerancia para el error

# Llamamos al método de Newton-Raphson pasando la función y su derivada
metodoN_R(f1, f_prime, x0, maX, espa)



```

```python
def steffensen(g, x0, eps, max_iter, valor_real):
    print(f"{'Iter':>4} | {'x_n':>12} | {'Error absoluto':>16}")
    print("-" * 36)
    
    iter = 0
    x1 = g(x0)
    
    while iter <= max_iter and abs(x1 - x0) > eps:
        x2 = g(x1)
        
        # Aplicar la fórmula de Steffensen
        num = (x1 - x0) ** 2
        denom = g(x2) - 2 * x1 + x0
        x_new = x1 - num / denom
        
        # Error absoluto
        error = abs(x_new - valor_real)
        print(f"{iter:4d} | {x_new:12.7f} | {error:16.10f}")
        
        iter += 1
        x0 = x1
        x1 = x_new

    if iter > max_iter:
        print("No converge en", max_iter, "iteraciones.")
    else:
        print(f"\nRaíz aproximada después de {iter} iteraciones: {x1:.7f}")
raiz_real = 0.6180339

steffensen(g7, x0=1, eps=0.0001, max_iter=20, valor_real=raiz_real)
#concluyo que el metodo esta divergiendo del resultado esperado
```
